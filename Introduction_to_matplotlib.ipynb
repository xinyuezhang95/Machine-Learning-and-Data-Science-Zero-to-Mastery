## matplotpib example workflow

```python
# 0. import matplotlib and get it ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. setup plot
fig, ax = plt.subplots(figsize = (10, 10)) # figsize = (width, height); one figure might contain several axes

# 3. plot data
ax.plot(x, y)

# 4. customize plot
ax.set(title = 'Simple Plot', xlabel = 'x-axis', ylabel = 'y-axis')

# 5. save & show (you save the figure)
fig.savefig('images/sample-plot.png')
```

## Two options for subplots

```python
# subplot option 1 (preferred in the videos)
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows = 2, ncols = 2, figsize = (10, 5))

# plot to each different axis
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
# subplot option 2
fig, ax = plt.subplots(nrows = 2, ncols = 2, figsize = (10, 5))

# plot to each different index
ax[0, 0].plot(x, x/2);
ax[0, 1].scatter(np.random.random(10), np.random.random(10));
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1, 1].hist(np.random.randn(1000));
```

### Which one should you use? (pyplot vs matplotlib OO method)

* When plotting something quickly, okay to use the pyplot method
* When plotting something more advanced, use the OO method

```python
heart_disease.head()
```

```python
# now let's plot with a more flexible way (Object Oriented plot)
over_50 = heart_disease[heart_disease['age'] > 50]
over_50.head()
```

```python
len(over_50)
```

```python
# pyplot method (aka plot directly from the pandas dataframe)
over_50.plot(kind = 'scatter', x = 'age', y = 'chol', c = 'target', figsize = (10, 6)); # c is short for color
```

```python
# OO method mixed with pyplot
fig, ax = plt.subplots(figsize = (10, 6))
over_50.plot(kind = 'scatter', x = 'age', y = 'chol', c = 'target',
            ax = ax); # now we have the 'age' axis
# this is actually the combination of OO and pyplot
```

```python
# OO method
fig, ax = plt.subplots(figsize = (10, 6))
over_50.plot(kind = 'scatter', x = 'age', y = 'chol', c = 'target',
            ax = ax); # now we have the 'age' axis
ax.set_xlim([45, 100]);
```

```python
# OO method from scratch
fig, ax = plt.subplots(figsize = (10, 6))

# plot the data
scatter = ax.scatter(x = over_50['age'], y = over_50['chol'], c = over_50['target'])

# customize the plot
ax.set(title = 'Heart Disease and Cholesterol Levels', 
      xlabel = 'Age',
      ylabel = 'Cholesterol')

# add a legend
ax.legend(*scatter.legend_elements(), title = 'Target'); 
### scatter.legend_elements(): go to the scatter variable, then find the legend elements, which is the c parameter, 
### then looks back to over_50['target'] column and finds its values

# add a horinzontal line
ax.axhline(over_50['chol'].mean(), linestyle = '--');
```

```python
over_50.head()
# thalach: max heart rate
```

```python
# subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows = 2, ncols = 1, figsize = (10, 10))

# add data to ax0
scatter = ax0.scatter(x = over_50['age'], y = over_50['chol'], c = over_50['target'])

# customize ax0
ax0.set(title = 'Heart Disease and Cholestorol Levels', xlabel = 'Age', ylabel = 'Cholestorol')

# add a legend to ax0
ax0.legend(*scatter.legend_elements(), title = 'Target')

# add a mean line
ax0.axhline(y = over_50['chol'].mean(), linestyle = '--');

# add data to ax1
scatter = ax1.scatter(x = over_50['age'], y = over_50['thalach'], c = over_50['target'])

# customize ax1
ax1.set(title = 'Heart Disease and Max Heart Rate', xlabel = 'Age', ylabel = 'Max Heart Rate')

# add a legend to ax1
ax1.legend(*scatter.legend_elements(), title = 'Target')

# add a mean line
ax1.axhline(y = over_50['thalach'].mean(), linestyle = '--');
```

```python
# similar plot, but with shared x axis

# subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows = 2, ncols = 1, figsize = (10, 10),
                              sharex = True)

# add data to ax0
scatter = ax0.scatter(x = over_50['age'], y = over_50['chol'], c = over_50['target'])

# customize ax0
ax0.set(title = 'Heart Disease and Cholestorol Levels', ylabel = 'Cholestorol')

# add a legend to ax0
ax0.legend(*scatter.legend_elements(), title = 'Target')

# add a mean line
ax0.axhline(y = over_50['chol'].mean(), linestyle = '--');

# add data to ax1
scatter = ax1.scatter(x = over_50['age'], y = over_50['thalach'], c = over_50['target'])

# customize ax1
ax1.set(title = 'Heart Disease and Max Heart Rate', xlabel = 'Age', ylabel = 'Max Heart Rate')

# add a legend to ax1
ax1.legend(*scatter.legend_elements(), title = 'Target')

# add a mean line
ax1.axhline(y = over_50['thalach'].mean(), linestyle = '--');

# add a title to the entire figure
fig.suptitle('Heart Disease Analysis', fontsize = 16, fontweight = 'bold');
```

```python
# only changes cmap

# set the style
plt.style.use('seaborn-whitegrid')

# OO method from scratch
fig, ax = plt.subplots(figsize = (10, 6))

# plot the data
scatter = ax.scatter(x = over_50['age'], y = over_50['chol'], c = over_50['target'],
                    cmap = 'summer') # this changes the color scheme

# customize the plot
ax.set(title = 'Heart Disease and Cholesterol Levels', 
      xlabel = 'Age',
      ylabel = 'Cholesterol')

# add a legend
ax.legend(*scatter.legend_elements(), title = 'Target'); 
### scatter.legend_elements(): go to the scatter variable, then find the legend elements, which is the c parameter, 
### then looks back to over_50['target'] column and finds its values

# add a horinzontal line
ax.axhline(over_50['chol'].mean(), linestyle = '--');
```

```python
# customizing the y and x axis limitations

# similar plot, but with shared x axis

# subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows = 2, ncols = 1, figsize = (10, 10),
                              sharex = True)

# add data to ax0
scatter = ax0.scatter(x = over_50['age'], y = over_50['chol'], c = over_50['target'],
                     cmap = 'winter')

# customize ax0
ax0.set(title = 'Heart Disease and Cholestorol Levels', ylabel = 'Cholestorol')
# change the x axis
ax0.set_xlim([50, 80]) # previously there's a line at x = 50

# add a legend to ax0
ax0.legend(*scatter.legend_elements(), title = 'Target')

# add a mean line
ax0.axhline(y = over_50['chol'].mean(), linestyle = '--');

# add data to ax1
scatter = ax1.scatter(x = over_50['age'], y = over_50['thalach'], c = over_50['target'],
                     cmap = 'winter')

# customize ax1
ax1.set(title = 'Heart Disease and Max Heart Rate', xlabel = 'Age', ylabel = 'Max Heart Rate')
# change ax1 axislimits
ax1.set_xlim([50, 80])
ax1.set_ylim([60, 200])

# add a legend to ax1
ax1.legend(*scatter.legend_elements(), title = 'Target')

# add a mean line
ax1.axhline(y = over_50['thalach'].mean(), linestyle = '--');

# add a title to the entire figure
fig.suptitle('Heart Disease Analysis', fontsize = 16, fontweight = 'bold');
```
